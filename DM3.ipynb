```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml import Pipeline

# Initialize Spark session
spark = SparkSession.builder.appName("EarthquakeModel").getOrCreate()

# Assuming earthquake_data_df is your Spark DataFrame
features = ['magnitude', 'depth', 'latitude', 'longitude']
target = 'number_of_buildings_impacted'

# VectorAssembler combines the feature columns into a single vector column
assembler = VectorAssembler(inputCols=features, outputCol="features")
data = assembler.transform(earthquake_data_df)

# Split the data into training and testing sets
train_data, test_data = data.randomSplit([0.8, 0.2], seed=42)

# StandardScaler to standardize the features
scaler = StandardScaler(inputCol="features", outputCol="scaled_features")

# Models
linear_model = LinearRegression(featuresCol="scaled_features", labelCol=target)
ridge_model = LinearRegression(featuresCol="scaled_features", labelCol=target, regParam=0.01, elasticNetParam=0)
lasso_model = LinearRegression(featuresCol="scaled_features", labelCol=target, regParam=0.01, elasticNetParam=1)

# Create pipelines
linear_pipeline = Pipeline(stages=[scaler, linear_model])
ridge_pipeline = Pipeline(stages=[scaler, ridge_model])
lasso_pipeline = Pipeline(stages=[scaler, lasso_model])

# Models list
models = [('Linear Regression', linear_pipeline), 
          ('Ridge Regression', ridge_pipeline), 
          ('Lasso Regression', lasso_pipeline)]

# Results dictionary
results = {}

# Evaluator for Mean Absolute Error (MAE)
evaluator = RegressionEvaluator(labelCol=target, predictionCol="prediction", metricName="mae")

# Evaluate each model
for model_name, model_pipeline in models:
    model_fitted = model_pipeline.fit(train_data)
    predictions = model_fitted.transform(test_data)
    mae = evaluator.evaluate(predictions)
    results[model_name] = {'MAE': mae}

results

```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.regression import LinearRegression
from pyspark.ml import Pipeline
import numpy as np
import matplotlib.pyplot as plt

# Initialize Spark session
spark = SparkSession.builder.appName("EarthquakeModel").getOrCreate()

# Read the data
earthquake_data_df = spark.read.csv("Dataset/reduced_data.csv", header=True, inferSchema=True)

features = ['magnitude', 'depth', 'latitude', 'longitude']
target = 'number_of_buildings_impacted'

# Split the data into training and testing sets
train_data, test_data = earthquake_data_df.randomSplit([0.8, 0.2], seed=42)

# Define the pipeline
assembler = VectorAssembler(inputCols=features, outputCol="features")
scaler = StandardScaler(inputCol="features", outputCol="scaled_features", withStd=True, withMean=False)
linear_model = LinearRegression(featuresCol="scaled_features", labelCol=target)
pipeline = Pipeline(stages=[assembler, scaler, linear_model])

# Fit the pipeline to training data
model = pipeline.fit(train_data)

# Define a range of latitudes for predictive analysis
latitudes = [float(i) for i in np.linspace(
    float(earthquake_data_df.agg({"latitude": "min"}).collect()[0][0]), 
    float(earthquake_data_df.agg({"latitude": "max"}).collect()[0][0]), 
    100
)]

predictive_df = spark.createDataFrame([
    (latitude, 
     earthquake_data_df.agg({"longitude": "mean"}).collect()[0][0], 
     earthquake_data_df.agg({"depth": "mean"}).collect()[0][0], 
     earthquake_data_df.agg({"magnitude": "mean"}).collect()[0][0]
    ) for latitude in latitudes
], features)

# Transform and predict using the model
predictions = model.transform(predictive_df)

predictions.select("prediction").describe().show()


# Extract predictions and plot
predicted_impacts = [row['prediction'] for row in predictions.select("prediction").collect()]

plt.figure(figsize=(10,6))
plt.plot(latitudes, predicted_impacts, color='navy', label='Predicted Impacts')
plt.title('Predicted Number of Buildings Impacted Across Different Latitudes')
plt.xlabel('Latitude')
plt.ylabel('Predicted Number of Buildings Impacted')
plt.axvline(x=40.57, color='r', linestyle='--', label='Latitude 40.57')
plt.axvline(x=-0.02, color='g', linestyle='--', label='Latitude -0.02')
plt.axvline(x=13.51, color='purple', linestyle='--', label='Latitude 13.51')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
