```python
from pyspark.sql import SparkSession
import matplotlib.pyplot as plt
import seaborn as sns

# Create a Spark session
spark = SparkSession.builder.appName("PCA Analysis").getOrCreate()

# Load the dataset
data = spark.read.csv('Dataset/reduced_data.csv', header=True, inferSchema=True)

# Descriptive statistics
desc_stats = data.describe().toPandas()

# Convert Spark DataFrame to Pandas for visualization
data_pd = data.toPandas()

# Visualization
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))

# Plot histograms
sns.histplot(data=data_pd, x='magnitude', kde=True, ax=axes[0, 0])
axes[0, 0].set_title('Magnitude Distribution')
sns.histplot(data=data_pd, x='depth', kde=True, ax=axes[0, 1])
axes[0, 1].set_title('Depth Distribution')
sns.histplot(data=data_pd, x='latitude', kde=True, ax=axes[0, 2])
axes[0, 2].set_title('Latitude Distribution')
sns.histplot(data=data_pd, x='longitude', kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Longitude Distribution')
sns.histplot(data=data_pd, x='number_of_buildings_impacted', kde=True, ax=axes[1, 1])
axes[1, 1].set_title('Number of Buildings Impacted Distribution')
sns.histplot(data=data_pd, x='economic_loss', kde=True, ax=axes[1, 2])
axes[1, 2].set_title('Economic Loss Distribution')

plt.tight_layout()
plt.show()

print(desc_stats)
```

```python
from pyspark.sql.functions import col
import pandas as pd
import numpy as np

# Filter out non-numeric columns
numeric_columns = [col_name for col_name, dtype in data.dtypes if dtype != 'string']

# Calculate the correlation matrix
n = len(numeric_columns)
correlation_matrix = np.zeros((n, n))
for i in range(n):
    for j in range(n):
        correlation_matrix[i, j] = data.stat.corr(numeric_columns[i], numeric_columns[j])

# Convert the matrix to a Pandas DataFrame for visualization
corr_df = pd.DataFrame(correlation_matrix, columns=numeric_columns, index=numeric_columns)

# Visualization
plt.figure(figsize=(10, 8))
sns.heatmap(corr_df, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()

```

```python
from pyspark.sql import SparkSession
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder

# Initialize Spark session
spark = SparkSession.builder.appName("TsunamiPrediction").getOrCreate()

# Assemble features using VectorAssembler
assembler = VectorAssembler(inputCols=['magnitude', 'depth', 'latitude', 'longitude'], outputCol="features")
df_assembled = assembler.transform(data)

# Split the data
(train_data, test_data) = df_assembled.randomSplit([0.8, 0.2], seed=42)

# Define the model
dt = DecisionTreeClassifier(labelCol="tsunami", featuresCol="features", seed=42)

# Parameter grid
paramGrid = ParamGridBuilder() \
    .addGrid(dt.maxDepth, [5, 10, 15, 20, 25]) \
    .addGrid(dt.minInstancesPerNode, [2, 5, 10, 20]) \
    .build()


# Cross validation
crossval = CrossValidator(estimator=dt,
                          estimatorParamMaps=paramGrid,
                          evaluator=MulticlassClassificationEvaluator(labelCol="tsunami", metricName="accuracy"),
                          numFolds=5)

# Train model
cvModel = crossval.fit(train_data)

# Get the best model and parameters
bestModel = cvModel.bestModel
bestParams = bestModel.extractParamMap()

# Get the best accuracy (optional: you would need to run the evaluator on a validation set)
evaluator = MulticlassClassificationEvaluator(labelCol="tsunami", predictionCol="prediction", metricName="accuracy")
bestAccuracy = evaluator.evaluate(bestModel.transform(test_data))

print(bestParams, bestAccuracy)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator, BinaryClassificationEvaluator

# Predict using the best model
predictions = bestModel.transform(test_data)

accuracy_evaluator = MulticlassClassificationEvaluator(
    labelCol="tsunami", predictionCol="prediction", metricName="accuracy"
)
accuracy_dt = accuracy_evaluator.evaluate(predictions)

f1_evaluator = MulticlassClassificationEvaluator(
    labelCol="tsunami", predictionCol="prediction", metricName="f1"
)
f1_dt = f1_evaluator.evaluate(predictions)

print(f"Accuracy: {accuracy_dt}")
print(f"F1 Score: {f1_dt}")
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Initialize Spark's logistic regression
lr = LogisticRegression(featuresCol="features", labelCol="tsunami")

# Set up the parameter grid
paramGrid_lr = ParamGridBuilder().addGrid(lr.regParam, [0.001, 0.01, 0.1, 1, 10, 100]).addGrid(lr.elasticNetParam, [0, 1]).build()

# Define evaluator
evaluator = BinaryClassificationEvaluator(rawPredictionCol="rawPrediction", labelCol="tsunami", metricName="areaUnderROC")

# Initialize cross-validator
crossval_lr = CrossValidator(estimator=lr,
                             estimatorParamMaps=paramGrid_lr,
                             evaluator=evaluator,
                             numFolds=5)

# Fit model
cvModel_lr = crossval_lr.fit(train_data)

# Extract the best model, parameters, and accuracy
bestModel_lr = cvModel_lr.bestModel
bestParams_lr = bestModel_lr.extractParamMap()
bestAccuracy_lr = evaluator.evaluate(bestModel_lr.transform(test_data))

print(bestParams_lr)
print(f"Best Accuracy: {bestAccuracy_lr}")
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# Predict the target variable for the test set
predictions_lr = bestModel_lr.transform(test_data)

# Evaluate the model
accuracy_evaluator_lr = MulticlassClassificationEvaluator(labelCol="tsunami", predictionCol="prediction", metricName="accuracy")
f1_evaluator_lr = MulticlassClassificationEvaluator(labelCol="tsunami", predictionCol="prediction", metricName="f1")

accuracy_lr = accuracy_evaluator_lr.evaluate(predictions_lr)
f1_lr = f1_evaluator_lr.evaluate(predictions_lr)

print(f"Accuracy (Logistic Regression): {accuracy_lr}")
print(f"F1 Score (Logistic Regression): {f1_lr}")
```

```python
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder

spark = SparkSession.builder.appName("EarthquakePrediction").getOrCreate()

assembler = VectorAssembler(inputCols=['latitude', 'longitude', 'magnitude', 'depth'], outputCol="features")
df_assembled = assembler.transform(data)

(train_data, test_data) = df_assembled.randomSplit([0.8, 0.2], seed=42)

# Initialize the Decision Tree Classifier
dt = DecisionTreeClassifier(labelCol="tsunami", featuresCol="features", seed=42)

# Parameter grid for hyperparameter tuning (you can adjust these parameters based on your needs)
paramGrid = ParamGridBuilder() \
    .addGrid(dt.maxDepth, [5, 10, 15]) \
    .addGrid(dt.minInstancesPerNode, [2, 5, 10]) \
    .build()

# Cross validation for model selection
crossval = CrossValidator(estimator=dt,
                          estimatorParamMaps=paramGrid,
                          evaluator=MulticlassClassificationEvaluator(labelCol="tsunami", metricName="accuracy"),
                          numFolds=5)

cvModel = crossval.fit(train_data)
bestModel = cvModel.bestModel

# Predict the target for the test set
predictions = bestModel.transform(test_data)

# Evaluate the model using accuracy and F1 score
evaluator_acc = MulticlassClassificationEvaluator(labelCol="tsunami", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator_acc.evaluate(predictions)

evaluator_f1 = MulticlassClassificationEvaluator(labelCol="tsunami", predictionCol="prediction", metricName="f1")
f1 = evaluator_f1.evaluate(predictions)

# Fetch feature importances from the model
feature_importances = bestModel.featureImportances.toArray()

# Combine feature names and their importance scores
features = ['latitude', 'longitude', 'magnitude', 'depth']
feature_importance_dict = dict(zip(features, feature_importances))

# Sort the features based on importance
sorted_feature_importance = sorted(feature_importance_dict.items(), key=lambda item: item[1], reverse=True)

print("Accuracy:", accuracy)
print("F1 Score:", f1)
print("Feature Importances:", sorted_feature_importance)

```
