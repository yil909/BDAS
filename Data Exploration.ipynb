```python
import findspark
findspark.init('/home/ubuntu/spark-3.2.1-bin-hadoop2.7')
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('basics').getOrCreate()
```

```python
file_path = 'Dataset/earthquake_data.csv'
df_earthquake = spark.read.csv(file_path, header=True, inferSchema=True)
num_rows = df_earthquake.count()
num_columns = len(df_earthquake.columns)
print(f"Number of Rows: {num_rows}")
print(f"Number of Columns: {num_columns}")
```

```python
df_earthquake.show(5)
```

```python
data_types = df_earthquake.dtypes
for column, dtype in data_types:
    print(f"{column}: {dtype}")
```

```python
key_attributes = ['magnitude', 'depth', 'cdi', 'mmi', 'sig', 'dmin', 'gap', 'longitude', 'latitude']

# Check if the key attributes are in the dataframe columns
missing_attributes = [attr for attr in key_attributes if attr not in df_earthquake.columns]

if not missing_attributes:
    # Compute basic statistics for key attributes
    basic_statistics = df_earthquake.select(key_attributes).describe().show()
    print("\nBasic Statistics:")
    print(basic_statistics)
else:
    print(f"\nThe following key attributes are missing in the dataset: {', '.join(missing_attributes)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Histograms
attributes = ['cdi', 'mmi', 'latitude', 'sig', 'depth', 'tsunami', 'longitude']
titles = ['Histogram for cdi', 'Histogram for mmi', 'Histogram for latitude', 'Histogram for sig', 
          'Histogram for depth', 'Histogram for tsunami', 'Histogram for longitude']

fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 20))
for ax, attr, title in zip(axes.flatten(), attributes, titles):
    sns.histplot(df_earthquake.select(attr).rdd.flatMap(lambda x: x).collect(), ax=ax, kde=True, bins=30)
    ax.set_title(title)
    ax.set_ylabel('Frequency')
fig.delaxes(axes[3, 1])
plt.tight_layout()
plt.show()

# Bubble plot for longitude, latitude, and magnitude
longitude = df_earthquake.select('longitude').rdd.flatMap(lambda x: x).collect()
latitude = df_earthquake.select('latitude').rdd.flatMap(lambda x: x).collect()
magnitude = df_earthquake.select('magnitude').rdd.flatMap(lambda x: x).collect()

plt.figure(figsize=(15, 10))
sns.scatterplot(x=longitude, y=latitude, size=magnitude, sizes=(10, 200), 
                hue=magnitude, palette='YlOrRd', alpha=0.6, edgecolor=None)
plt.title('Bubble plot for Longitude, Latitude, and Magnitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Magnitude', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Check for Missing Data
from pyspark.sql import functions as F
# Check for missing values in each column
missing_data = df_earthquake.select([F.sum(F.when(F.isnull(c), 1).otherwise(0)).alias(c) for c in df_earthquake.columns])

# Calculate the percentage of missing data
total_rows = df_earthquake.count()
missing_percentage = missing_data.select([(F.col(c) / total_rows * 100).alias(c) for c in df_earthquake.columns])

# Combine the missing data counts and percentages into one DataFrame
missing_overview = missing_data.union(missing_percentage)

# Display the results
missing_overview.show()
$
```

```python
# Defining categorical and continuous variables
categorical_vars = ['title', 'date_time', 'alert', 'net', 
                    'magType', 'location', 'continent', 'country']
continuous_vars = ['magnitude', 'cdi', 'mmi', 'tsunami', 'sig', 
                    'nst', 'dmin', 'gap', 'depth', 'latitude', 'longitude']

# Checking unique values and their counts for categorical variables
categorical_values = {}
for cat_var in categorical_vars:
    categorical_values[cat_var] = df_earthquake.groupBy(cat_var).count().collect()

# Checking the min, max, and mean values for continuous variables
continuous_stats = df_earthquake.describe(continuous_vars)

# Display the results
for cat_var, values in categorical_values.items():
    print(cat_var, values)
continuous_stats.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

numerical_columns = ['magnitude', 'depth', 'cdi', 'mmi']
numerical_df = df_earthquake.select(numerical_columns)

# Convert to pandas DataFrame for visualization
pdf = numerical_df.toPandas()

# Plotting the boxplots
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15, 10))
fig.suptitle('Distribution of Numerical Columns', fontsize=16, y=1.05)

for ax, column in zip(axes.flatten(), numerical_columns):
    sns.boxplot(y=pdf[column], ax=ax)
    ax.set_title(f'Distribution of {column}')
    ax.set_ylabel(column)

plt.tight_layout()
plt.show()
```
